## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for importing in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title = 'Sample Plot',
      xlabel = 'x-axis',
      ylabel = 'y-axis')

# 5. save and show (save the whole figure)
fig.savefig('images/sample-plot.png')
```

## Making figures with NumPy arrays

* line plot
* scatter plot
* bar plot
* histogram
* subplot

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data as a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# plot the data as a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# plot data from dictionary as a bar plot
nut_butter_prices = {
    'Almond butter': 10,
    'Peanut butter': 8,
    'Cashew butter': 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = 'Nut Butter Store',
      ylabel = 'Price ($)');
```

```python
# plot data from dictionary as a horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title = 'Nut Butter Store',
      xlabel = 'Price ($)');
```

```python
# make data and plot as histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# plot data as subplot (option 1)
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))

# plot to each axis
ax0.plot(x, x/2);
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.hist(np.random.randn(1000));
```

```python
# plot data as subplot (option 2)
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

# plot to each index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which method to use? pyplot vs. matplotlib OO method

* when plotting quickly, pyplot is good
* when plotting more advanced data, use matplotlib OO 

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target');
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

ax.set(title = 'Heart Disease and Cholesterol Levels',
       xlabel = 'Age',
       ylabel = 'Cholesterol');

ax.legend(*scatter.legend_elements(), title = 'Target');

ax.axhline(over_50['chol'].mean(),
           linestyle = '--');
```

```python
plt.style.use('seaborn-whitegrid')

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10),
                               sharex = True)

# chol
scatter0 = ax0.scatter(x = over_50['age'],
                       y = over_50['chol'],
                       c = over_50['target'],
                       cmap = 'winter')

ax0.set(title = 'Heart Disease and Cholesterol Levels',
       ylabel = 'Cholesterol');

ax0.set_xlim([50, 80])

ax0.legend(*scatter0.legend_elements(), title = 'Target');

ax0.axhline(over_50['chol'].mean(),
           linestyle = '--');

#thalach
scatter1 = ax1.scatter(x = over_50['age'],
                       y = over_50['thalach'],
                       c = over_50['target'],
                       cmap = 'winter')

ax1.set(title = 'Heart Disease and Maximum HR',
       xlabel = 'Age',
       ylabel = 'Maximum HR');

ax1.set_ylim([60, 200])

ax1.legend(*scatter1.legend_elements(), title = 'Target');

ax1.axhline(over_50['thalach'].mean(),
           linestyle = '--');

fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
